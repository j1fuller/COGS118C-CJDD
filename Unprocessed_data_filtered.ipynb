```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.stats import ttest_ind
import mne

# Define File Paths and Subject IDs for AD and Control
ad_files = [
    "/Users/cooperward/Downloads/sub-008_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-002_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-003_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-004_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-007_task-eyesclosed_eeg-unprocessed.set"
]
ad_subject_ids = ["Subject 008", "Subject 002", "Subject 003", "Subject 004", "Subject 007"]

control_files = [
    "/Users/cooperward/Downloads/sub-037_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-042_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-039_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-044_task-eyesclosed_eeg-unprocessed.set",
    "/Users/cooperward/Downloads/sub-041_task-eyesclosed_eeg-unprocessed.set"
]
control_subject_ids = ["Subject 037", "Subject 042", "Subject 039", "Subject 044", "Subject 041"]

# Preprocessing Function: Band-pass filter, ICA
def preprocess_eeg(file_path, l_freq=0.5, h_freq=45.0, n_components=0.95):
    """
    Loads an EEGLAB .set file, sets a standard 10-20 montage, 
    applies a band-pass filter (0.5â€“45 Hz), 
    creates a synthetic EOG channel from Fp1 and Fp2, 
    runs ICA to remove EOG artifacts, 
    and returns the cleaned Raw object.
    """
    # Load raw data
    raw = mne.io.read_raw_eeglab(file_path, preload=True)

    # Set a standard 10-20 montage
    montage = mne.channels.make_standard_montage("standard_1020")
    raw.set_montage(montage)

    # Create a synthetic EOG channel from Fp1 and Fp2
    eog_data = raw.copy().pick_channels(["Fp1", "Fp2"]).get_data().mean(axis=0, keepdims=True)
    info_eog = mne.create_info(["EOG"], sfreq=raw.info["sfreq"], ch_types=["eog"])
    eog_raw = mne.io.RawArray(eog_data, info_eog)
    raw.add_channels([eog_raw])
    print("Synthetic EOG channel created from Fp1 and Fp2.")

    # Apply a band-pass filter
    raw.filter(l_freq=l_freq, h_freq=h_freq, fir_design="firwin")
    print(f"Band-pass filter applied: {l_freq}â€“{h_freq} Hz.")

    # Run ICA
    ica = mne.preprocessing.ICA(n_components=n_components, random_state=97, max_iter="auto")
    ica.fit(raw)

    # Find EOG-related components
    eog_indices, _ = ica.find_bads_eog(raw, ch_name="EOG", threshold=3.0)
    ica.exclude = eog_indices
    print("Identified EOG-related ICA components:", eog_indices)

    # Apply ICA to remove those components
    raw_clean = ica.apply(raw.copy())
    return raw_clean

# Plot EEG Time Series
def plot_eeg_time_series(eeg_data, subject_id, selected_channels=[0,1,2,14,15],
                         channel_names=["Fp1", "Fp2", "F3", "T5", "T6"],
                         colors=["r", "g", "b", "y", "k"], fs=500, time_limit=20, ax=None):
    """
    Plots the EEG time series for selected channels.
    """
    n_samples = eeg_data.shape[1]
    time = np.linspace(0, n_samples/fs, n_samples)
    
    if ax is None:
        fig, ax = plt.subplots(figsize=(12, 5))
        
    for i, ch_idx in enumerate(selected_channels):
        ax.plot(time, eeg_data[ch_idx, :]*1e6, label=channel_names[i], color=colors[i])
    ax.set_xlim(0, time_limit)
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("EEG Signal (ÂµV)")
    ax.set_title(f"EEG Time Series for {subject_id}")
    ax.legend()
    ax.grid(True)

# PSD Computation Using Welch's Method
def compute_psd_welch(eeg_data, fs=500, nperseg=1024):
    """
    Computes the PSD for each channel using Welch's method.
    Returns frequency bins and PSD array of shape (n_channels, n_freqs).
    """
    n_channels = eeg_data.shape[0]
    psd_list = []
    for ch in range(n_channels):
        freqs, pxx = signal.welch(eeg_data[ch, :]*1e6, fs=fs, nperseg=nperseg)
        psd_list.append(pxx)
    return freqs, np.array(psd_list)

# Relative Band Power Calculation Function
def compute_relative_power(psd, band_mask, broad_mask):
    """
    Computes relative power as (sum of PSD in the band)/(sum of PSD in the broad range)
    for each channel, then averages across channels.
    """
    total_power = np.sum(psd[:, broad_mask], axis=1)
    band_power = np.sum(psd[:, band_mask], axis=1)
    rel_power = np.divide(band_power, total_power, out=np.zeros_like(band_power), where=total_power!=0)
    return np.mean(rel_power)

# Preprocess, Plot Before/After, Compute PSD, Relative Power, Stats
fs = 500  # Sampling frequency

# Containers for processed data
ad_processed_data = []
control_processed_data = []

# For each AD subject, plot raw vs. processed side by side.
for fp, sid in zip(ad_files, ad_subject_ids):
    raw_unproc = mne.io.read_raw_eeglab(fp, preload=True)
    data_unproc = raw_unproc.get_data()
    
    raw_clean = preprocess_eeg(fp, l_freq=0.5, h_freq=45.0, n_components=0.95)
    data_proc = raw_clean.get_data()
    
    fig, axs = plt.subplots(1, 2, figsize=(16, 5))
    plot_eeg_time_series(data_unproc, subject_id=sid + " (Before)", ax=axs[0])
    axs[0].set_ylim(-500, 500)
    plot_eeg_time_series(data_proc, subject_id=sid + " (After)", ax=axs[1])
    axs[1].set_ylim(-200, 200)
    plt.suptitle(f"EEG Time Series for {sid}: Before vs. After Processing", fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
    
    ad_processed_data.append(data_proc)

# For each Control subject, do the same.
for fp, sid in zip(control_files, control_subject_ids):
    raw_unproc = mne.io.read_raw_eeglab(fp, preload=True)
    data_unproc = raw_unproc.get_data()
    
    raw_clean = preprocess_eeg(fp, l_freq=0.5, h_freq=45.0, n_components=0.95)
    data_proc = raw_clean.get_data()
    
    fig, axs = plt.subplots(1, 2, figsize=(16, 5))
    plot_eeg_time_series(data_unproc, subject_id=sid + " (Before)", ax=axs[0])
    axs[0].set_ylim(-500, 500)
    plot_eeg_time_series(data_proc, subject_id=sid + " (After)", ax=axs[1])
    axs[1].set_ylim(-200, 200)
    plt.suptitle(f"EEG Time Series for {sid}: Before vs. After Processing", fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
    
    control_processed_data.append(data_proc)

# Compute PSD for Each Subject (Processed Data)
ad_psd_list = []
control_psd_list = []
frequencies = None

for data in ad_processed_data:
    f, psd = compute_psd_welch(data, fs=fs)
    if frequencies is None:
        frequencies = f
    ad_psd_list.append(psd)

for data in control_processed_data:
    f, psd = compute_psd_welch(data, fs=fs)
    control_psd_list.append(psd)

print("PSD computed for all subjects.")
print("Frequency bins shape:", frequencies.shape)

# Define Frequency Masks and Compute Relative Band Power
theta_mask = (frequencies >= 4) & (frequencies <= 7)
alpha_mask = (frequencies >= 8) & (frequencies <= 12)
gamma_mask = (frequencies >= 30) & (frequencies <= 45)
broad_mask = (frequencies >= 1) & (frequencies <= 50)

ad_theta_rel = [compute_relative_power(psd, theta_mask, broad_mask) for psd in ad_psd_list]
ad_alpha_rel = [compute_relative_power(psd, alpha_mask, broad_mask) for psd in ad_psd_list]
ad_gamma_rel = [compute_relative_power(psd, gamma_mask, broad_mask) for psd in ad_psd_list]

control_theta_rel = [compute_relative_power(psd, theta_mask, broad_mask) for psd in control_psd_list]
control_alpha_rel = [compute_relative_power(psd, alpha_mask, broad_mask) for psd in control_psd_list]
control_gamma_rel = [compute_relative_power(psd, gamma_mask, broad_mask) for psd in control_psd_list]


# T-tests on Relative Power
t_theta, p_theta = ttest_ind(ad_theta_rel, control_theta_rel)
t_alpha, p_alpha = ttest_ind(ad_alpha_rel, control_alpha_rel)
t_gamma, p_gamma = ttest_ind(ad_gamma_rel, control_gamma_rel)

print("\nRelative Power T-test Results:")
print(f"Theta: t = {t_theta:.3f}, p = {p_theta:.3f}")
print(f"Alpha: t = {t_alpha:.3f}, p = {p_alpha:.3f}")
print(f"Gamma: t = {t_gamma:.3f}, p = {p_gamma:.3f}")

# Group Comparison Visualization: Bar Plot with Error Bars
bands = ["Theta", "Alpha", "Gamma"]
ad_means = [np.mean(ad_theta_rel), np.mean(ad_alpha_rel), np.mean(ad_gamma_rel)]
ctrl_means = [np.mean(control_theta_rel), np.mean(control_alpha_rel), np.mean(control_gamma_rel)]
ad_stds = [np.std(ad_theta_rel), np.std(ad_alpha_rel), np.std(ad_gamma_rel)]
ctrl_stds = [np.std(control_theta_rel), np.std(control_alpha_rel), np.std(control_gamma_rel)]

x = np.arange(len(bands))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, ad_means, width, yerr=ad_stds, capsize=5, label="AD")
plt.bar(x + width/2, ctrl_means, width, yerr=ctrl_stds, capsize=5, label="Control")
plt.xticks(x, bands)
plt.ylabel("Relative Power (Band / Total)")
plt.title("Relative Band Power by Group")
plt.legend()
plt.tight_layout()
plt.show()
```
